```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv") #data read
df
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price ($US)')
plt.scatter(df.area, df.price, color ='red', marker='+') #data plot
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']], df.price) # data train LR_Model
```

```python
reg.predict([[3300]]) # model prediction
```

```python
reg.coef_ # model coefficent
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color ='red', marker='+') #data plot
plt.plot(df.area,reg.predict(df[['area']]), color='blue')
```
